```python
import cv2
import numpy as np
```

```python
canvas = np.zeros((300, 300, 3), dtype='uint8')
```

```python
cv2.imshow('canvas',canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Line

```python
pt1 = (0,0)
pt2 = (300, 300)
color = (0,255,0)
cv2.line(canvas, pt1, pt2, color)

cv2.imshow('cavas', canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
pt1 = (0, 300)
pt2 = (300, 0)
cv2.line(canvas, pt1, pt2, (0,0,255), 5)

cv2.imshow('cavas', canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Rectangle

```python
canvas = np.zeros((600,600,3), dtype='uint8')
cv2.imshow('canvas',canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
pt1 = (50, 50)
pt2 = (100, 100)
color = (255, 0, 0)

cv2.rectangle(canvas, pt1, pt2, color)
cv2.rectangle(canvas, (100,100), (200,200), (0,255,0),5)
cv2.rectangle(canvas, (200,200), (300,300), (0,255,0),-1)

cv2.imshow('canvas', canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Polygon

```python
canvas =255 * np.ones((600, 600, 3), dtype='uint8')
cv2.imshow('canvas',canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# points of hexagon
pts = np.array([[200,200],[250,300],[350,300], [400, 200], [350,100], [250,100]])
pts = pts.reshape(-1,1,2)

cv2.polylines(canvas, [pts], True, (255,0,0))
cv2.imshow('canvas', canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Circles

```python
canvas =250 * np.ones((600,900,3), dtype='uint8')
cv2.imshow('canvas',canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

cv2.circle(canvas, (100,200),100, (255,0,0), 5)
cv2.circle(canvas, (400,200),100, (0,255,0), 5)
cv2.circle(canvas, (700,200),100, (0,0,255), 5)
cv2.circle(canvas, (250,300),100, (255,0,255), 5)
cv2.circle(canvas, (550,300),100, (255,255,0), 5)
cv2.imshow('canvas',canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Text

```python
cv2.putText(canvas, 'Olympic', (200, 500), cv2.FONT_HERSHEY_PLAIN, 5, (0,0,0), 5)
cv2.imshow('canas',canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
