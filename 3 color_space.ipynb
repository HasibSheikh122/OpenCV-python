Convert Color

```python
img1 = cv2.imread('./images/beach.jpg')
```

```python
cv2.imshow('orginal-bgr', img1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# BGR to Gray
gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

# BGR to RGB
img_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

# BGR to HSV
img_hsv = cv2.cvtColor(img1, cv2.COLOR_BGR2HSV)


# BGR to LAB
img_lab = cv2.cvtColor(img1, cv2.COLOR_BGR2Lab)

cv2.imshow('orginal image', img1)
cv2.imshow('Gray scale', gray)
cv2.imshow('RGB scale', img_rgb)
cv2.imshow('HSB scale', img_hsv)
cv2.imshow('LAB scale', img_lab)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
